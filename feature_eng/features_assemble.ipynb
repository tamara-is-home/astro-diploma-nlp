### Features description:

`refs_count : int` - represents the count of references and citations to other posts included in the telegram body. A higher value suggests that the post is more comprehensive and supported by a significant amount of evidence from external sources.

`telegram_len: int` - the length of the telegram body. The longer length might indicate more substantial observations.

`month: int [1-12]` - the number of the month when observations took place. Regarding to the data analysis we made before, some monthes (June, October) get more citations than the others (December, July).

#### Time-series features (each feature is calculated with respect to the time and never considers the target values from the current observation):
$Note$: by the `year` I mean the previous 365 days, by the `month` - the previous 30 days. Not the calendar month / year.

#### general features:
`month_citation_rate_to_global_ratio: float` -  the average number of citations this month divided by the global average number of citations, reflects the relative interest in recent posts. A higher value of this metric (above 1) suggests an increasing level of interest in the posts made during the current month compared to the overall average citation rate.

`year_citation_rate_to_global_ratio: float` - calculated as the average number of citations during the current year period divided by the global average number of citations. A higher value of this metric (above 1) suggests a growing level of interest in posts made thru the current year.

#### author-based features:

`author_all_time_citation_rate_to_global_ratio: float` - calculated as the average number of citations received by this author(s)' all previous publications divided by the global average number of citations. A value greater than 1 indicates that the author may be relatively more popular or well-known compared to other authors.

`author_year_citation_rate_to_global_ratio: float` - calculated as the average number of citations received by the author's posts during the current year and dividing it by the global average number of citations during the same period.

`author_month_citation_rate_to_global_ratio: float` - calculated as the average number of citations received by the author's posts during the current month and dividing it by the global average number of citations during the same period.

`author_activity_frac_year: float` - the ratio of the number of posts by the author during the current year to the total number of posts by this author. A lower value (minimum of 0) indicates that this is the author's first post in the current year, suggesting limited activity. A higher value (maximum of 1) indicates that all of the author's activity occurred in the current year, suggesting that the author may be relatively new to the field or have a focused presence within a specific time frame.

`author_activity_frac_month: float` - the ratio of the number of posts by the author during the current month to the total number of posts by this author. Same as the feature mentioned above but for the month time period.

#### topics-based features:

`topics_all_time_citation_rate_to_global_ratio: float` - computed as the average number of citations received by all previous publications that share the same topics as the current one, divided by the global average number of citations. A value greater than 1 suggests that these specific topics, in combination, have a higher level of interest and citation compared to other topics. It indicates that publications within these topics tend to receive more attention and citations in general.

`topics_year_citation_rate_to_global_ratio: float` - computed as the average number of citations received by previous publications posted this year that share the same topics as the current one, divided by this year global average number of citations. A value greater than 1 suggests that these specific topics have a higher level of interest during the current year than the other topics.

`topics_month_citation_rate_to_global_ratio: float` - computed as the average number of citations received by previous publications posted this month that share the same topics as the current one, divided by this month global average number of citations.  A value greater than 1 suggests that these specific topics have a higher level of interest during the current month than the other topics.

`topics_activity_frac_year` - the ratio of the number of posts with the same topics during the current year to the total number of posts with these topics. A lower value (minimum of 0) indicates that these topics are not being actively disscussed this year. A higher value (maximum of 1) indicates that most of these topics' related posts occurred in the current year, suggesting that these topics might either be more common for a specific time frame or they are getting more popular recently. 

`topics_activity_frac_month` - same as the feature mentioned above but for the month time period.

`topics_weight_coef_all_time: float` - the weigh coefficient of the topics which takes into the consideration their frequency during the whole time. For each topic we calculate it's relative importance (weight) as:
$$ 
\frac{1}{number\_of\_obseravtions\_with\_this\_topic} * \frac{total\_number\_of\_observations}{number\_of\_unique\_topics}
$$

By applying weights to the citation rates of the topics, we can highlight the less frequent topics by normalizing their citation rates based on their respective weights. For example, topics such as 'transient' or 'optical' are commonly observed in many posts and therefore have weights of 0.1 and 0.2, while the topic 'meteor' is rare and has a weight of 400. The overall weight coefficient is calculated by dividing the raw citations by the weighted citations. A higher value indicates the presence of rare topics.

`topics_weight_coef_year`  - represents the topics weights computed using the mentioned above approach but within a yearly time window. A higher value indicates the presence of topics that are observed less frequently compared to others within this year.

`topics_weight_coef_month` - represents the topics weights computed using the same approach but within a monthly time window. A higher value indicates the presence of topics that are observed less frequently compared to others within this month.
