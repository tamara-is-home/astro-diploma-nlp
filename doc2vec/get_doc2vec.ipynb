```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/assembled.csv", index_col=0)
```

### read the doc2vecC representations

```python
vectors = []
with open("doc2vecC/docvectors.txt", 'r') as f:
    for line in f.readlines():
        vectors.append(line)
```

```python
len(vectors)  # one extra vector (all zeros) - empty newline at the end of the training file
```

```python
np_vectors = []
for vector in vectors[:-1]:
    np_vectors.append(np.array([np.float64(el) for el in vector.split(' ') if el != '\n']))
```

```python
df['doc_vec'] = np_vectors
df.head()
```

```python
# save the data and proceed
pd.DataFrame(df['doc_vec'].to_list(), index=df.index).to_csv("doc_vectors.csv", index=True)
```
